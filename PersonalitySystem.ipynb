### Introduction

##### The objective of this project is to build a personality prediction system using the Big Five Personality Test dataset provided on Kaggle, collected (2016-2018) through an interactive on-line personality test. (Link: https://www.kaggle.com/datasets/tunguz/big-five-personality-test/data)

##### We aim to build a system to classify personality traits based on these user responses to the self-descriptive questions asked by the personality test.

1. import necessary libraries

```python
import pandas as pd
import numpy as np
from sklearn.cluster import MiniBatchKMeans
```

2. read dataframe from file and display

```python
df = pd.read_csv("IPIP-FFM-data-8Nov2018/data-final.csv", delimiter = "\t")
df
```

3. Explore dataframe

```python
# Print columns
columns = df.columns
for column in columns:
    print(column)
```

```python
# Extract the first 50 columns 
x = df[df.columns[0:50]]
# Removes the limit of the number of columns shown
x
```

4. Check for missing values

```python
print("Missing value in data? ", "Yes" if x.isna().any().any() else "No")
```

5. Initialize a MiniBatchKMeans clustering model, and fit it into x

```python
kmeans = MiniBatchKMeans(n_clusters = 16, random_state = 0, batch_size = 100, max_iter=100, n_init='auto').fit(x)
```

```python
# Check the number of cluster centers/personality types
len(kmeans.cluster_centers_)
```

```python
# Create a dictionary to map personality types to cluster centers
personality_clusters = {f"Type {i+1}": center for i, center in enumerate(kmeans.cluster_centers_)}

# Access a specific cluster
print("Cluster for Type 1: \n\n", personality_clusters["Type 1"])
```
